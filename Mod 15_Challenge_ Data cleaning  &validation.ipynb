```python
import math

def find_median(list_of_things):
    length_list = []
    
    for thing in list_of_things:
        length_list.append(thing)
    
    length_list.sort()
    count = len(length_list)
    isEven = count % 2 == 0
  
    if (isEven):
        mid = math.floor( count / 2 )
        a = length_list[mid - 1]
        b = length_list[mid]
        ans = (a + b) / 2
    else:
        ans = length_list[math.floor( count / 2 )]
        
    return ans
```
